```python
#Parse HTML Content using BeatifulSoup

#Unstructured
import requests
from bs4 import BeautifulSoup

url = 'https://courses.wscubetech.com/'
r = requests.get(url)
#print(r)
#print(r.text)

#structured

soup = BeautifulSoup(r.text,'lxml')
print(soup)

```

```python
#The BeautifulSoup find() function

url = 'https://webscraper.io/test-sites/e-commerce/allinone/computers/tablets'
r = requests.get(url)
#print(r)

soup = BeautifulSoup(r.text,'lxml')
print(soup.find('div'))
```

```python
url = 'https://webscraper.io/test-sites/e-commerce/allinone/computers/tablets'
r = requests.get(url)
#print(r)

soup = BeautifulSoup(r.text,'lxml')
price = (soup.find('h4',{'class':'float-end price card-title pull-right'}))
print(price.string)
```

```python
soup = BeautifulSoup(r.text,'lxml')
price = (soup.find('h4',class_ = 'float-end price card-title pull-right'))
print(price.string)
```
